```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("my-app") \
    .master("local[*]") \
    .getOrCreate()
```

```python
spark
```

```python
from pyspark.sql.functions import col
```

```python
df = spark.range(10).toDF("id")
df = df.withColumn("doble", col("id") * 2)
df.explain()
```
